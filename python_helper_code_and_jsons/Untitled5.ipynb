```python
import os
import shutil
```

```python
docs = "../docs"
public = "../public"
```

```python
os.listdir(public)
```

```python
publics = ['crh.png','poster.png','scientist.png']
docs_files = os.listdir(docs)
for file in publics:
    if file not in docs_files:
        shutil.copy(os.path.join(public, file), os.path.join(docs, file))
```

```python
directory = "../docs/assets"
files = [i for i in os.listdir(directory) if i[-3:]==".js"]

changes = [("/sign_in.html", "../sign_in.html"),
          ("protect_data.html", "../protect_data.html"),
          ("upload.html", "../upload.html"),
          ("./src/upload.html", "../src/upload.html"), 
          ("./about.html", "../about.html"),]

for file in files:
    path = os.path.join(directory, file)
    with open(path) as f:
        text = f.read()
    for old,new in changes:
        text = text.replace(old,new)
    with open(path, "w") as f:
        f.write(text)
```
