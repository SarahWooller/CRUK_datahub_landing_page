```python
import os
import json
```

```python
dire  = "../src/utils/new_dummies"
```

```python
os.listdir(dire)
```

```python
for file in os.listdir(dire):
    num = file[8:10]
    path  = os.path.join(dire,file)
    with open(path) as f:
        j = json.load(f)
        for i,grant in enumerate (j["projectGrants"]):
            grant["pid"] = f"{num}_{str(i).zfill(3)}"
        print(j["projectGrants"])
    with open(path, "w") as f:
        json.dump(j, f, indent=3)
```
