```python
import mammoth
import os

def convert_docx_to_md(input_path, output_path):
    """
    Converts a .docx file to a .md file using mammoth.
    """
    if not os.path.exists(input_path):
        print(f"Error: The file '{input_path}' does not exist.")
        return

    with open(input_path, "rb") as docx_file:
        # Perform the conversion
        result = mammoth.convert_to_markdown(docx_file)
        markdown_content = result.value
        messages = result.messages

        # Write the converted content to the output file
        with open(output_path, "w", encoding="utf-8") as md_file:
            md_file.write(markdown_content)

        # Output any warnings generated during conversion
        if messages:
            print("Conversion notes:")
            for msg in messages:
                print(msg)
        else:
            print(f"Conversion of '{input_path}' to '{output_path}' is complete.")
```

```python
def generate_markdown_table(raw_data):
    """
    Converts a tab-separated string into a Markdown table.
    """
    # Split the raw data into rows and cells, ignoring completely empty lines
    lines = [line for line in raw_data.strip().split('\n') if line.strip()]
    if not lines:
        return "No data provided."

    rows = [line.split('\t') for line in lines]

    # Find the maximum width for each column to keep the Markdown aligned
    col_widths = []
    num_columns = max(len(row) for row in rows)
    
    for col_idx in range(num_columns):
        max_width = max(
            len(str(row[col_idx]).strip()) if col_idx < len(row) else 0 
            for row in rows
        )
        col_widths.append(max(max_width, 3)) # Minimum width of 3 for '---'

    md_table = []

    def format_row(row_data):
        cells = []
        for i in range(num_columns):
            # Handle empty or missing cells
            val = str(row_data[i]).strip() if i < len(row_data) else ""
            cells.append(val.ljust(col_widths[i]))
        return "| " + " | ".join(cells) + " |"

    # 1. Generate Header
    md_table.append(format_row(rows[0]))

    # 2. Generate Separator (e.g., |---|---|)
    separator = ["-" * width for width in col_widths]
    md_table.append("|-" + "-|-".join(separator) + "-|")

    # 3. Generate Data Rows
    for row in rows[1:]:
        md_table.append(format_row(row))

    return "\n".join(md_table)
```
