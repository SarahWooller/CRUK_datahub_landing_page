```python
with open("filter_information.json") as f:
    filters = json.load(f)
```

```python
filter_ids = [flter['id'] for flter in filters]
```

```python
def get_mock_study_list():
    studies_list = list(range(100))
    random.shuffle(studies_list)
    return studies_list[:random.randint(30,50)]
```

```python
mock_study_list = dict([(flt,get_mock_study_list()) for flt in filter_ids])
```

```python
studies = dict([(i, list(set(map(lambda x : str(int(x)), np.random.randint(0, high=50, size=random.randint(30,50)))))) for i in filter_ids])
```

```python
with open("mock_study_data.js", "w") as f:
    f.write("export const studyData = ")
with open("mock_study_data.js", "a") as f:
    json.dump(studies, f)
with open("mock_study_data.js", "a") as f:
    f.write(";")
```
